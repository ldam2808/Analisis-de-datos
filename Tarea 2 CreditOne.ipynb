### Class Factor

De la figura anterior, ser mujer o soltera reduce la posibilidad de incumplimiento de pago.

### Age Factor

```python
#Linear plot of age vs. defaulted
sns.lmplot('AGE', 'default', data=credit)
```

Parece haber una ligera tendencia lineal entre la edad y el campo default. El gráfico muestra que cuanto más viejo es el cliente, más probabilidades hay de que no cumplan el pago.

```python
# Defaulted vs. Age grouped by gender
sns.lmplot('AGE', 'default', data=credit, hue='SEX')
```

Los hombres mayores tienen una mayor tasa de incumplimiento en el pago como se muestra en la figura anterior

```python
# Defaulted vs. Age grouped by Martial Status
sns.lmplot('AGE', 'default', hue='MARRIAGE', data=credit)
```

Los clientes divorciados tienen la mayor probabilidad de incumplir un pago a medida que envejecen. Los solteros tienen una probabilidad ligeramente menor de incumplir un pago a medida que envejecen.
